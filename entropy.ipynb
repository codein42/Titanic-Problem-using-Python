```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total = 100

li = []

for red in range(1, 100):
    yellow = total - red
    
    p_red = red/total
    p_yellow = yellow/total
    
    entropy = -(p_red*np.log2(p_red) + p_yellow*np.log2(p_yellow))
    
    li.append(entropy)
```

```python
plt.plot(li)
```
